```python
from datasets import load_dataset
```

```python
dataset = load_dataset("MLNTeam-Unical/OpenTuringBench", name="in_domain")
```

```python
df_train = dataset["train"].to_pandas()

df_test = dataset["test"].to_pandas()
```

```python
mapping_classes = {
    "meta-llama": "Meta",
    "qwen": "OpenAI",
    "mistralai": "Mistral",
    "google": "Google",
    "anthropic": "Anthropic",
}
```

```python
df_train["model_family"] = df_train["model"].apply(lambda x: mapping_classes.get(x.split("/")[0].lower(), "Others"))
```

```python
df_test["model_family"] = df_test["model"].apply(lambda x: mapping_classes.get(x.split("/")[0].lower(), "Others"))
```
